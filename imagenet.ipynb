```python
import os

import numpy as np

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
from torch.optim import Adam

import torchvision
from torchvision import transforms
```

```python
#@title Model pretrained on imagenet

os.environ["TORCH_MODEL_ZOO"] = "./data/pretrained_models"
model = torchvision.models.resnet18(pretrained=True)
```

```python
##@title Testing model with random images on internet

from PIL import Image
import urllib.request
urllib.request.urlretrieve("https://media.istockphoto.com/photos/chicken-isolated-on-white-picture-id148127611?k=20&m=148127611&s=612x612&w=0&h=xQmE1alpIqvVGp9DcvLV8UJ--_xrPauc4Z5R588kZTg=", "img.jpg")
transform = transforms.Compose([
 transforms.Resize(256),
 transforms.CenterCrop(224),
 transforms.ToTensor(),
 transforms.Normalize(
 mean=[0.485, 0.456, 0.406],
 std=[0.229, 0.224, 0.225]
 )])
img = transform(Image.open("img.jpg")).unsqueeze(0)
model.eval()
logits = model(img)
prob = F.softmax(logits, dim=1)
print(f"Class: {prob.argmax()}, Prob: {prob.max()}")
```
