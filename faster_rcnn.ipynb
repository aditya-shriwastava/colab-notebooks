```python
import voc_utils
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
def get_voc_data(dataset, n):
    xmin = int(dataset.loc[n].xmin)
    ymin = int(dataset.loc[n].ymin)
    xmax = int(dataset.loc[n].xmax)
    ymax = int(dataset.loc[n].ymax)
    bbox = np.array([xmin, ymin, xmax, ymax])

    # float32 image with values between [0,1]
    img = voc_utils.load_img(dataset.loc[n].fname.split(".")[0])
    return img, bbox
    
def disp_data(img, bbox):    
    # From float32 image to uint8 image
    img = (img * 255).astype(np.uint8)
    
    cv2.rectangle(img , bbox[0:2], bbox[2:4],(0,255,0),2)
    plt.imshow(img)
    plt.show()
    print(f"Image Shape: {img.shape}")
    print(f"BBOX: [xmin: {bbox[0]}, ymin: {bbox[1]}, xmax: {bbox[2]}, ymax: {bbox[3]}]")
```
