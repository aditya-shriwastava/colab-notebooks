```python
def get_board(lines):
  board = []
  for line in lines:
    board.append([int(x) for x in line.split()])
  return board 

with open('input_p4.txt', 'r') as fd:
  lines = fd.readlines()
  draws = [int(draw) for draw in lines[0].split(',')]
  boards = []
  for i in range(len(lines[1:])//6):
    board = get_board(lines[(i*6 + 2):(i*6 + 7)])
    boards.append(board)
  boards = np.array(boards)

print(draws)
print(boards[:2])
```

```python
def has_won(board, draws):
  has_won = False

  # Mark board_bool based on draws
  board_bool = np.array([value in draws for value in board.reshape(-1)]).reshape(5,5)

  # Check rows and cols for victory
  for i in range(5):
    if board_bool[i,:].all() or board_bool[:,i].all():
      has_won = True
      break
  return has_won

def get_score(board, draws):
  board_bool = np.array(
    [value in draws for value in board.reshape(-1)]
  ).reshape(5,5)
  board_score = board[np.invert(board_bool)].sum() * draws[-1]
  return board_score
```

```python
# Finding Winner board
winner_board = None
winner_draws = None
for l in range(len(draws)):
  for board in boards:
    if has_won(board, draws[:l]):
      winner_board = board
      winner_draws = draws[:l]
      break
  if winner_board is not None:
    break

print(winner_board)
print(winner_draws)
print(f"winner_board_score: {get_score(winner_board, winner_draws)}")
```

```python
with open('input_p5.txt', 'r') as fd:
  line_segments = []
  for line in fd.readlines():
    start, end = line.split('->')
    x1,y1 = start.split(',')
    x2,y2 = end.split(',')
    line_segments.append([int(x1), int(y1), int(x2), int(y2)])
  line_segments = np.array(line_segments)
```

```python
xmax = max(line_segments[:,0].max(), line_segments[:,2].max())
ymax = max(line_segments[:,1].max(), line_segments[:,3].max())

grid = np.zeros((xmax+1, ymax+1))
for line_segment in line_segments:
  x1, y1, x2, y2 = line_segment
  if y1 == y2: # Horizontal Line Segment
    grid[y1, min(x1,x2):max(x1,x2)+1] += 1
  elif x1 == x2: # Vertical Line Segment
    grid[min(y1,y2):max(y1,y2)+1, x1] += 1
  
print(grid)
print(f"Result: {(grid>=2).sum()}")
```
